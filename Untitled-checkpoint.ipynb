```python
df = pd.read_csv("test_data.csv", low_memory = False)
```

Calucating additional features for analysis

```python
df['ROI(%)'] = ((df['Profit']-df['Balance'])/df['Balance'])*100
#df['ROI(%)']= np.where((df['Profit']>0, temp , 0)
df['Open_Datetime'] = pd.to_datetime(df['OpenTime'],unit='s')
df['Open_month'] = pd.DatetimeIndex(df['Open_Datetime']).month
df['Open_day'] = pd.DatetimeIndex(df['Open_Datetime']).day
df['New_Equity'] = df['Balance'] + df['Profit']
df['Close_Datetime'] = pd.to_datetime(df['CloseTime'],unit='s')
df['Close_month'] = pd.DatetimeIndex(df['Open_Datetime']).month
df['Close_day'] = pd.DatetimeIndex(df['Close_Datetime']).day
df['Close_year'] = pd.DatetimeIndex(df['Close_Datetime']).year
df['Inactive_days'] = (pd.Timestamp.now() - df['Close_Datetime']).dt.days
#df['win_loss'] = (df['TradeAction'] == 3) & (df['Profit'] > 0)  #(+ve profit) true for all closed, (-ve or open) -> false
df = df.replace([np.inf, -np.inf], np.nan) #remove inf value
df.fillna(0) #fill NAN with 0's
```

Divide the above profiles into 3 quartiles to segmentize on performance. Defining functions. 

```python
def getQ(df):
    quantiles = df.quantile(q=[0.33,0.66])
    q = quantiles.to_dict()
    return q

def roi_score(q, ROI): #1 for least ROI, 3 for most ROI 
    grp = []
    for i in range(len(ROI)):
        if ROI[i] <= q['ROI(%)'][0.33]:
            grp.append(1)
        elif ROI[i] <= q['ROI(%)'][0.66] and ROI[i] >= q['ROI(%)'][0.33]:
            grp.append(2)
        else:
            grp.append(3)
    return grp

def active_score(q, active): # 1 for most active, 3 for least active
    grp = []
    for i in range(len(active)):
        if active[i] <= q['Inactive_days'][0.33]:
            grp.append(3)
        elif active[i] <= q['Inactive_days'][0.66] and active[i] >= q['ROI(%)'][0.33]:
            grp.append(2)
        else:
            grp.append(1)
    return grp
```

Convert profile features to lists for easy comparisons.

```python
troi = traderP['ROI(%)'].values.tolist()
tact = traderP['Inactive_days'].values.tolist()
croi = cityP['ROI(%)'].values.tolist()
cact = cityP['Inactive_days'].values.tolist()
coroi = countryP['ROI(%)'].values.tolist()
coact = countryP['Inactive_days'].values.tolist()
iroi = instrumentP['ROI(%)'].values.tolist()
mroi = monthP['ROI(%)'].values.tolist()
yroi = yearP['ROI(%)'].values.tolist()
```

Call the function to calculate score for each

```python
traderP['rscore'] = roi_score(getQ(traderP),troi)
traderP['iscore'] = active_score(getQ(traderP),tact)
cityP['rscore'] = roi_score(getQ(cityP),croi)
cityP['iscore'] = active_score(getQ(cityP),cact)
countryP['rscore'] = roi_score(getQ(countryP),coroi)
countryP['iscore'] = active_score(getQ(countryP),coact)
instrumentP['rscore'] = roi_score(getQ(instrumentP),iroi)
monthP['rscore'] = roi_score(getQ(monthP),mroi)
yearP['rscore'] = roi_score(getQ(yearP),yroi)
traderP['tscore'] = traderP['rscore'].map(str) + traderP['iscore'].map(str) # 33 being the highest score
cityP['cscore'] = cityP['rscore'].map(str) + cityP['iscore'].map(str)
countryP['coscore'] = countryP['rscore'].map(str) + countryP['iscore'].map(str)
print(traderP)
```
